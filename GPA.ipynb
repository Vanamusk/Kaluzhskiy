Script to translate GPA and make plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

Function to translate U.S. grade system into Israel grade system (data collected from [here](https://www.scholaro.com/db/countries/Israel/Grading-System)):

```python
y_fit = np.array([60,70,80,90])
x_fit = np.array([2,2.3,3,4])
f = np.poly1d(np.polyfit(x_fit,y_fit, deg = 2))
```

GPA:

```python
x =  np.array([1,2,3,4,5,6,7]) #semester nembers
y = np.array([4.88,4.75,5.25,5.45,6.0,4.45,6.89]) #GPA per semester in MIPT grade system

y_alt = np.array([2.63, 2.63, 2.75, 2.82, 3, 2.55, 3.44]) #GPA per semester in U.S. grade system
y_t = [[3,2,4,3,2,2,3,2], #all grades in U.S. grade system
        [3,2,3,2,3,4,2,2],
        [2,3,4,2,3,2,4,2],
        [3,3,3,3,2,2,2,3,3,4,3],
        [2,3,3,2,3,3,3,4,3,3,4],
        [2,2,3,2,4,3,2,3,2,2],
        [3,4,2,4,4,3,3,4,4]] 
cred_t = [[3,6,1,3,6,6,3,4], #credits
          [2,3,4,6,6,3,4,6],
          [3,2,4,3,2,3,3,5],
          [2,4,3,1,4,3,5,5,1,3,4],
          [3,2,5,4,1,2,2,1,3,2,3],
          [3,3,3,1,1,2,1,3,3,5],
          [1,2,3,2,2,3,3,12,1]]
y_cum = np.array([sum(np.array(g)*np.array(c))/sum(np.array(c))for g,c in zip(y_t,cred_t)]) #cummulative U.S. GPA
```

plots:

```python
plt.figure(figsize = (15,5))
plt.subplot(1,3,1,)
plt.plot(x,y, label = 'GPA')
plt.title('10-point MIPT grade system')
#plt.suptitle('Grade', x = 0.02, y = 0.5, rotation='vertical')
#plt.xlabel('Number of semester')
plt.ylabel('Grade')
plt.legend()
plt.subplot(1,3,2)
plt.plot(x,y_alt, label = 'GPA')
plt.plot(x,y_cum, label = 'cummulative GPA')
plt.title('4-point U.S. grade system')
plt.suptitle('Semesters number', y = 0.03)
plt.legend()
plt.subplot(1,3,3)
plt.title('100-point Israel grade system')
plt.plot(x,f(y_alt), label = 'GPA')
plt.legend();
```
